### 1. Select customer name together with each order the customer made

```python
import pandas as pd
query1 = '''SELECT CustomerName, OrderID
            FROM customers c
            JOIN orders o
            ON c.CustomerID = o.CustomerID;
        '''
pd.read_sql_query(query1, conn)
```

### 2. Select order id together with name of employee who handled the order

```python
query2 = '''SELECT o.OrderID, e.EmployeeID, e.FirstName, e.LastName
            FROM orders o
            JOIN employees e
            ON o.EmployeeID = e.EmployeeID'''

pd.read_sql_query(query2, conn)
```

### 3. Select customers who have not placed any order yet

```python
query3 = '''SELECT c.CustomerID, c.CustomerName, o.OrderID
            FROM customers c
            LEFT JOIN orders o
            ON c.CustomerID = o.CustomerID
            WHERE o.CustomerID IS NULL'''

pd.read_sql_query(query3, conn)
```

### 4. Select order id together with the name of products


```python
query4 = '''SELECT o.OrderID, p.ProductID, p.ProductName
            FROM orders o
            JOIN order_details od ON o.OrderID = od.OrderID
            JOIN products p ON p.ProductID = od.ProductID
            ORDER BY o.OrderID'''

pd.read_sql_query(query4, conn)
```

### 5. Select products that no one bought

```python
query5 = '''SELECT p.ProductID, p.ProductName, od.OrderID
            FROM products p
            LEFT JOIN order_details od ON p.ProductID = od.ProductID
            WHERE od.OrderID IS NULL'''

pd.read_sql_query(query5, conn)
```

### 6. Select customer together with the products that he bought

```python
query6 = '''SELECT c.CustomerID, c.CustomerName, p.ProductName
            FROM customers c
            JOIN orders o ON o.CustomerID = c.CustomerID
            JOIN order_details od ON od.OrderID = o.OrderID
            JOIN products p ON p.ProductID = od.ProductID'''

pd.read_sql_query(query6, conn)
```

### 7. Select orders together with the name of the shipping company

```python
query7 = '''SELECT o.OrderID, o.CustomerID, o.EmployeeID, o.OrderDate, shp.ShipperName
            FROM orders o
            JOIN shippers shp
            ON o.ShipperID = shp.ShipperID
            ORDER BY o.OrderID'''

pd.read_sql_query(query7, conn)
```

### 8. Select customers with id greater than 50 together with each order they made

```python
query8 = '''SELECT c.CustomerID, c.CustomerName, o.OrderID
            FROM customers c
            JOIN orders o
            ON c.CustomerID = o.CustomerID
            WHERE c.CustomerID > 50'''

pd.read_sql_query(query8, conn)
```

### 9. Select employees together with orders with order id greater than 1040

### 10. Select the most expensive product

```python
query10 = '''SELECT ProductID,ProductName,Price
            FROM products
            ORDER BY Price DESC
            LIMIT 1'''

pd.read_sql_query(query10, conn)
```

### 11. Select the second most expensive product

```python
query11 = '''SELECT ProductID,ProductName,Price
            FROM products
            ORDER BY Price DESC
            LIMIT 1 OFFSET 1;'''

pd.read_sql_query(query11, conn)
```

### 12. Select name of the cheapest product (only name) without using LIMIT and OFFSET


```python
query12 = '''with temp as
            (SELECT ProductName,MIN(Price)
            FROM products)
            SELECT ProductName
            FROM
            temp;'''

pd.read_sql_query(query12, conn)
```

### 15. Select customer name together with the number of orders made by the corresponding customer and sort the result by number of orders in decreasing order


```python
query15 = '''SELECT c.CustomerID, c.CustomerName, COUNT(*) AS Totalorder
            FROM customers c
            JOIN Orders o
            ON c.CustomerID = o.CustomerID
            GROUP BY c.CustomerID
            ORDER BY 3 DESC'''
pd.read_sql_query(query15, conn)
```

### 16. Select the customer who spent the most money

```python

query16 = '''SELECT c.CustomerID, c.CustomerName, SUM(od.Quantity * p.Price) AS TotalSpent
            FROM orders o
            JOIN customers c ON o.CustomerID = c.CustomerID
            JOIN order_details od ON o.OrderID = od.OrderID
            JOIN products p ON p.ProductID = od.ProductID
            GROUP BY c.CustomerID
            ORDER BY 3 DESC
            LIMIT 1;'''

pd.read_sql_query(query16, conn)
```

### 17. Select the customer who spent the second most money and lives in Canada

```python
query17 = '''SELECT c.CustomerID, c.CustomerName, SUM(od.Quantity * p.Price) AS TotalSpent, c.Country
            FROM orders o
            JOIN customers c ON o.CustomerID = c.CustomerID
            JOIN order_details od ON o.OrderID = od.OrderID
            JOIN products p ON p.ProductID = od.ProductID
            WHERE c.Country = 'Canada'
            GROUP BY c.CustomerID
            ORDER BY 3 DESC
            LIMIT 1 OFFSET 1;'''

pd.read_sql_query(query17, conn)
```

### 18.Select shipper together with the total price of processed orders

```python

query18 = '''SELECT o.ShipperID, shp.ShipperName, SUM(od.Quantity * p.Price) AS TotalOrderValue
            FROM orders o
            JOIN order_details od ON o.OrderID = od.OrderID
            JOIN products p ON p.ProductID = od.ProductID
            JOIN shippers shp ON shp.ShipperID = o.ShipperID
            GROUP BY 1
            ORDER BY 2;'''

pd.read_sql_query(query18, conn)
```
